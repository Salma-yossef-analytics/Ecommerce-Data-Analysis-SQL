```python
# 1. إعداد الاتصال بقاعدة البيانات
# سيتم إنشاء الملف في نفس مجلد المشروع
conn = sqlite3.connect("ecommerce.db")

```

```python
# 2. قراءة البيانات من ملف CSV
# latin1 لمعالجة الرموز الخاصة
path = 'C:/Users/PLUS VAIO/OneDrive/سطح المكتب/projects/ecommerce/data.csv'
df = pd.read_csv(path, encoding='latin1')

```

 الطريقة الأولى: التنظيف قبل الرفع (Pre-processing) ---


```python

# 3. خطوة التنظيف  (Data Cleaning)
# تحويل عمود التاريخ من نص  Datetime Object قبل رفعه للقاعدة
#  وبالتالي ستتعامل معها دوال التجميع على أنها نص 
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

```

```python
# 4. نقل البيانات إلى SQL
# if_exists='replace' تضمن تحديث البيانات في كل مرة يتم فيها تشغيل الكود
df.to_sql("ecommerce", conn, if_exists='replace', index=False)

print("--- تم إنشاء قاعدة البيانات وتحويل التواريخ بنجاح ---")

```

```python
# 5. الاستعلام عن أقدم وأحدث فاتورة باستخدام  
#  الاستعلام الآن بسيط وسريع
query = """
SELECT 
    MIN(InvoiceDate) AS Oldest_Invoice,
    MAX(InvoiceDate) AS Latest_Invoice
FROM ecommerce;
"""

result = pd.read_sql_query(query, conn)

```

```python
# 6. عرض النتائج النهائية
print("\nنتائج تحليل المواعيد:")
print(result)

```

```python
# إغلاق الاتصال بالقاعدة ( للحفاظ على الموارد)
conn.close()
```

 الطريقة الثانية: التحويل بعد سحب البيانات من القاعدة (Post-processing) ---
تستخدم في حال كانت البيانات مخزنة كنص ولانملك صلاحية تعديل الجدول 

```python
# (هذه الطريقة التي اعتمدناها في الكود بالأعلى بوضع pd.to_datetime قبل to_sql)


# 1. إعادة فتح الاتصال (لأغراض العرض)
conn = sqlite3.connect("ecommerce.db")

# 2. سحب البيانات الخام (Raw Data)
df_raw = pd.read_sql_query("SELECT InvoiceDate FROM ecommerce", conn)

# 3. تحويل النوع داخل الباندا بعد الاستعلام
df_raw['InvoiceDate'] = pd.to_datetime(df_raw['InvoiceDate'])

# 4. استخراج النتائج
print("\nنتائج الحل الثاني (التحويل بعد القراءة):")
print(f"Oldest: {df_raw['InvoiceDate'].min()}")
print(f"Latest: {df_raw['InvoiceDate'].max()}")
```
